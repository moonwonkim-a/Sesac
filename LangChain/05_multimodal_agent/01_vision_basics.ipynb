## 환경설정

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
```

```python
image_url = "https://png.pngtree.com/png-vector/20250105/ourlarge/pngtree-anatomical-depiction-of-a-human-heart-with-lifelike-details-emphasizes-scientific-png-image_15051779.png"

model = ChatOpenAI(model='gpt-4o',temperature=0)

# 멀티모달 메세지 구성
# 텍스트와 이미지 URL을 함께 리스트로 전달
message = HumanMessage(
    content=[
        {"type" : "text", "text" : "이 이미지는 뭐야? 자세히 묘사해줘."},    # 질문
        {"type" : "image_url", "image_url" : {"url":image_url}}             # 이미지 데이터
    ]
)

response = model.invoke([message])      # 리스트에 묶어서 보냄
print(response.content)
```

```python
# 영수증 이미지
receipt_url = "https://img.etnews.com/photonews/1707/971120_20170705143932_354_0001.jpg"

prompt_text = "이 이미지에서 상호명(Store name)과 총 금액(Totall amount)을 찾아서 알려줘"

message = HumanMessage(
    content=[
        {"type" : "text", "text" : "이 이미지는 뭐야? 자세히 묘사해줘."},    # 질문
        {"type" : "image_url", "image_url" : {"url":receipt_url}}             # 이미지 데이터
    ]
)

r = model.invoke([message])
print(r.content)
```
