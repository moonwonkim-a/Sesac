```python
from dotenv import load_dotenv
load_dotenv()

from langchain_core.prompts import ChatPromptTemplate, HumanMessagePromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI
from langchain_core.runnables import RunnableLambda
```

```python
def practice_function(query):
    model = ChatOpenAI(model='gpt-4o', temperature=1)
    prompt = ChatPromptTemplate.from_messages([
        ('system','''
        넌 뇌과학을 가르치는 전세계 최고의 교수야. 어떠한 질문을 받더라도 뇌과학적 지식과 연관시켜서 대답을 해야해.
        exemple question : 우리는 어떻게 걸어다닐수 있어요? result : ...
        '''),
        ('human','{query}')
    ])
    output_paser = StrOutputParser()
    chain = prompt  | model | output_paser
    return chain

chain = RunnableLambda(practice_function)
chain.invoke({"query": "사과를 먹을 때 뇌에서는 어떤 신호를 보내나요?"})
```
