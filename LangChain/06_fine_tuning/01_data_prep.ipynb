```python
# 환경설정
from dotenv import load_dotenv
load_dotenv()

import json
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage
```

```python
model = ChatOpenAI(model="gpt-5-nano", temperature=0.9)
```

```python
# 데이터 생성 프롬프트
generation_prompt = """
넌 Fine_tuning을 위한 학습 데이터를 생성하는 데이터 엔지니어야.
내 목표는 '고객의 질문에 대해 트렌디하고, 친근하게 답하는 인스타그램 마케터 봇'을 만드는 거야.

아래 JSON 포맷에 맞춰서, 다양한 고객 질문과 그에 맞는 트렌디한 답변 쌍을 10개만 생성해줘.
주제는 [패션, 맛집, 여행, 운동] 등등 다양하게 해줘.

Format:
{'messages' : [{'role':'system', 'content':'너는 트렌디하고 친절한 인스타 마케터 봇이야.'}, {'role':'user', 'content':'질문'}, {'role':'assistant','content':'답변'}]}
{'messages': .... }

오직 JSONL 형식의 텍스트만 출력해줘. 코드 블록 없이
"""

response = model.invoke(generation_prompt)
print(response.content)
```
