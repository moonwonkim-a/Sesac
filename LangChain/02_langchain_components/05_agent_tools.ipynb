```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from langchain_classic.agents import AgentExecutor, create_tool_calling_agent, load_tools
from langchain_classic import hub

model = ChatOpenAI(
    model='gpt-5',
    temperature=0.2
)

# 도구 로드
# - 'wikipedia' : 위키백과 검색 도구
# - 'llm-math' : 계산기가 필요한 수학 문제를 풀 때 사용하는 도구
tools = load_tools(['wikipedia', 'llm-math'], llm=model)

# 도구 사용에 최적화된 표준 프롬프트를 가져온다.
prompt = hub.pull("hwchase17/openai-tools-agent")
```

```python
# 에이전트 생성
agent = create_tool_calling_agent(model, tools, prompt)

# 실행기 설정
agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True,               # 에이전트의 '생각의 흐름'을 우리 눈에 보여주는 옵션
    handle_parsing_errors=True  # 에러발생시 에이전트가 재시도하도록 설정
)
```

```python
model.invoke('혼성그룹 ALLDAY PROJECT 멤버를 알려줘')
```

```python
agent_executor.invoke({'input':'혼성그룹 ALLDAY PROJECT 멤버를 알려줘'})
```
