## 데이터 변환하기2

```python
import pandas as pd

df = pd.read_csv('./data/seoul_park02.csv')
df.head()
```

# 열 전체 변환 : map()

```python
week = {0:'월', 1:'화', 2:'수', 3:'목',4:'금', 5:'토', 6:'일'}
```

```python
df['요일']=df['요일'].map(week)
```

```python
df.head()
```

### 데이터에 함수 적용 : apply()

```python
df['날씨'].value_counts()
```

```python
# '눈', '비' 를 입력 받으면 '눈/비'로 바꿔주는 함수
def weather(e):
    if e == '눈' or e == '비':
        return '눈/비'
    else:
        return e
```

```python
df['날씨'] = df['날씨'].apply(weather)

df['날씨'].value_counts()
```

```python
# 람다 함수 적용
df['날씨'] = df['날씨'].apply(lambda e : '눈/비' if e == '눈' or e == '비' else e)
```
