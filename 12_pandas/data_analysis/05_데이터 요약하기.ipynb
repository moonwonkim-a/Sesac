## 데이터 요약하기

```python
import pandas as pd

df = pd.read_csv("./data/seoul_park03.csv")

df.head()
```

### 데이터 통계값 확인 : mean(), min(), max(), median()

```python
df['어른'].mean()   # 어른 컬럼의 평군값(전체 일자의 어른 입장객 수의 평균)
```

```python
df['총계'].min()    # 총 입장객 수가 가장 적은 날의 입장객 수
```

```python
# mm 변수에 미세먼지 데이터 불러와서 저장
mm = pd.read_csv("./data/misemunji.csv")

# 집계함수를 사용해 미세먼지가 가장 높은 날의 수치 변수를 badair에 저장
badair = mm['미세먼지'].max()
badair
```

### 데이터 그룹화 : groupby()

```python
# '날씨'를 기준으로 '총계'값의 평균 구하기
df.groupby('날씨')['총계'].mean()   # groupby('기준컬럼')['값으로 쓸 컬럼']
```

```python
df.groupby(['날씨','공휴일'])['총계'].mean()    # []로 두 개 이상 컬럼을 묶어서 사용 가능
```

```python
df.groupby(['날씨','공휴일'])[['어른','어린이']].mean()     # 기준컬럼은() 안에 []추가 // 값 컬럼은[] 안에 [] 추가
```

```python
# 목요일이면서 공휴일인 날짜의 어른 입장객 수 최대값을 변수 thursday에 저장
df.groupby(['요일','공휴일'])['어른'].max()
thursday = 38511
```
