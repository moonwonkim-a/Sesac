### 데이터 병합 : concat()

```python
# concat([합칠 데이터1, 합칠 데이터2], axis=0, join='inner',ignore_index=True)
# axis=0 -> 행으로 합치기// axis=1 -> 열로 합치기
# join='inner' -> 공통 컬럼들만 합쳐지고 나머지 컬럼들은 버림 // 'outer' -> 기본값, 모든 컬럼들 다 합치기
# ignore_index=True -> 기존 인덱스 무시하고 새로운 인덱스 만들기 // 기본값은 False로 되어있음
pd.concat([df,df2], axis=0, join='inner',ignore_index=True)
```

### 데이터 병합 : merge()

```python
mm = pd.read_csv('./data/misemunji.csv')
mm.head()
```

```python
mm['날짜'] = pd.to_datetime(mm['날짜'], format="%Y%m%d")    # 날짜 타입 맞추기
df["날짜"] = pd.to_datetime(df["날짜"], format="%Y%m%d")    # 날짜 타입 맞추기
```

```python
# on : 데이터프레임을 합칠 때 기준이 되는 컬럼의 이름
# how : 데이터 프레임의 데이터를 합치는 방법(inner, outer, left, right)
# inner : 교집합(두 데이터프레임이 모두 존재하는 데이터만)
# outer : 합칩합(두 데이터프레임의 모든 데이터 사용)
# left : 왼쪽 데이터프레임에 있는 모든 데이터
# right : 오른쪽 데이터프레임에 있는 모든 데이터
pd.merge(df, mm, on='날짜', how='left')     # merge(병합할 데이터1, 병합할 데이터2, on='기준컬럼', how='병합방식')
```

```python
score = pd.DataFrame({
    '이름': ['민수', '도윤', '서아', '연우', '하준'],
    '국어': [95, 72, 83, 89, 100],
    '수학': [88, 72, 93, 80, 95],
    '영어': [96, 68, 85, 94, 93]},
)

add1 = pd.DataFrame({
    '이름': ['지호', '준우', '예서'],
    '국어': [88, 92, 100],
    '수학': [97, 91, 100],
    '영어': [87, 95, 97],
    '역사': [95, 100, 92]}
)

add2 = pd.DataFrame({
    '이름': ['민수', '서아', '하준'],
    '과학': [93, 82, 89],
    '프로그래밍': [97, 96, 94]}
)
```

`score`에는 기존의 특별반 5명 학생들의 **이름**과 **국어, 수학, 영어** 점수가 담겨있습니다.
| | 이름  | 국어  | 수학  | 영어  |
|---|---|---|---|---|
| 0 | 민수  | 95  | 88  | 96  |
| 1 | 도윤  | 72  | 72  | 68  |
| 2 | 서아  | 83  | 93  | 85  |
| 3 | 연우  | 89  | 80  | 94  |
| 4 | 하준  | 100 | 95  | 93  |

`add1`에는 3명의 새로운 특별반 학생들의 **이름, 국어, 수학, 영어, 역사**점수가 기록되어 있습니다.

| | 이름  | 국어  | 수학  | 영어  | 역사 |
|---|---|---|---|---|---|
| 0 | 지호  | 88  | 97  | 87  | 95 |
| 1 | 준우  | 92  | 91  | 95  | 100 |
| 2 | 예서  | 100  | 100  | 97  | 92 |

`add2`에는 기존의 특별반 5명의 학생들 중 **과학**과 **프로그래밍**시험을 본 학생들의 시험점수가 **이름**과 함께 기록되어있습니다. 

| | 이름  | 과학  | 프로그래밍 |
|---|---|---|---|
| 0 | 민수  | 93  | 97  |
| 1 | 서아  | 82  | 96  |
| 2 | 하준  | 89  | 94  |

```python
# 1. score와 add1을 활용하여 전체 특별반 8명 학생들의 국어, 수학, 영어 점수데이터를 만들어 ans1에 저장 (hint .concat() 사용)
ans1 = pd.concat([score, add1],axis=0,join='inner',ignore_index=True)
ans1
```

```python
# 2. score와 add2를 활용하여 기존 특별반 학생들 중 국어, 수학, 영어, 과학, 프로그래밍 5개 과목의 시험을 모두 본 학생들의 점수 데이터를 만들어 ans2에 저장(hint .merge() 사용)
ans2 = pd.merge(score,add2, on='이름', how='right')
ans2
```
