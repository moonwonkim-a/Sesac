# 데이터 정제하기

```python
import pandas as pd

df = pd.read_csv("./data/seoul_park03.csv")
df.head()
```

### 데이터 정렬 : sort_values

```python
# 특정 컬럼의 값을 기준으로 전체 데이터를 정렬한다.
df_sorted = df.sort_values("총계", ascending = False)  # ascending 생략하면 기본값 True(오름차순)

df_sorted.head()
```

```python
df_sorted.tail(10)
```

```python
# 미세먼지 데이터에서 초미세먼지 농도가 3번째로 심했던 날짜를 pollution에 저장

mm = pd.read_csv('./data/misemunji.csv')
mm.sort_values('초미세먼지',ascending = False)
```

### 인덱스 재지정 : reset_index

```python
# 데이터 프레임의 인덱스를 처음부터 다시 지정해 줄 때 사용
# drop 옵션을 주지 않으면 기존의 index를 지정한 "index" 컬럼이 생김
df_sorted.reset_index(drop=True)    # drop을 True로 지정하여 "index" 컬럼 생성 안함
```

### 데이터 삭제 : drop

```python
# df.drop("행/열이름", axis=0 또는 1) 0이면 행 삭제, 1이면 열 삭제

# 삭제할 2개의 컬럼을 리스트로 입력
# axis=1로 설정하여 컬럼을 삭제
df=df.drop(["유료합계","무료합계"], axis=1)
df
```

### 열 이름 바꾸기 : rename

```python
df = df.rename(columns = {'총계': '총입장객수'})    # rename(columns={'기존컬럼명':'바꿀컬럼명'})
df
```

### 결측치 채우기 : fillna()

```python
df.fillna(int(df['청소년'].mean()))
```

### 결측치 데이터 삭제하기 : dropna()

```python
df.dropna(subset='청소년')
```

```python
df['청소년'] = df['청소년'].fillna(int(df['청소년'].mean()))
```

```python
# 미세먼지 데이터에서 결측치가 있는 컬럼을 파악하고 해당 컬럼의 중앙값(median())으로 채운 뒤 mm에 덮어씌워 저장

# 1. 미세먼지 데이터를 불러와 mm에 저장
mm = pd.read_csv('./data/misemunji.csv')
# 2. 어떤 컬럼에 결측치가 존재하는지 확인
mm.isnull().sum()
# mm의 결측치를 채워넣은 뒤 덮어 씌우자
mm['오존'] = mm['오존'].fillna(mm['오존'].median())
```
