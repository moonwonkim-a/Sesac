# 데이터 변환하기1

```python
import pandas as pd

# 파일 불러오기
df = pd.read_csv('./data/seoul_park.csv')
df.head()
```

```python
df
```

```python
df.info()
```

```python
# 텍스트 데이터를 정수형으로 바꾸기 전에 텍스트에서 숫자만 남기고 제거
# 이를 위해 숫자로 바꾸어야 하는 컬럼들에서 replace()를 활용하여 -를 0으로 바꾸고 콤마를 제거한다.
# str.replace()
colums = ['유료합계', '어른', '청소년', '어린이', '외국인', '단체', '무료합계', '총계']
for i in colums:
    df[i] = df[i].str.replace('-','0')  # 왼쪽의 값을 오른쪽의 값으로 바꾼다.
    df[i] = df[i].str.replace(',','')
```

## 데이터 타입 변환 : astype()
원하는 타입으로 변환가능

```python
df.info()   # 여전히 타입은 object로 되어있음
```

```python
df['어른'].astype(int)  # '어른' 컬럼을 astype을 이용하여 정수형(int)로 변환 // 원본에 반영되진 않았음
```

```python
df.info()
```

```python
df['어른'] = df['어른'].astype(int)     # 변환된 데이터 타입을 원본에 덮어 씌우는 과정이 필요하다.

df.info()
```

### 데이터 타입 변환 : to_numeric()
숫자로만 변환된다.

```python
df['유료합계']=pd.to_numeric(df['유료합계'])

df.info()
```

```python
# df의 "외국인" 컬럼의 자료형을 숫자형으로 변환해 변수 foreigner에 저장
foreigner = pd.to_numeric(df['외국인'])

foreigner.info()
```

```python
colums = ['유료합계', '어른', '청소년', '어린이', '외국인', '단체', '무료합계', '총계']
for i in colums:
    df[i] = pd.to_numeric(df[i])

df.info()
```

### 데이터 타입 변환 : to_datetime()

```python
df["날짜"] = pd.to_datetime(df["날짜"])
df.info()
```

```python
# datetime 형태의 데이터는 연,월,일 정보가 담겨있어 이를 활용하여 새로운 정보를 생성할 수 있다.
df['year'] = df['날짜'].dt.year
df['month'] = df['날짜'].dt.month
df['day'] = df['날짜'].dt.day
df['week'] = df['날짜'].dt.dayofweek    # 요일은 숫자로 저장(월~일 : 0~6)
```
