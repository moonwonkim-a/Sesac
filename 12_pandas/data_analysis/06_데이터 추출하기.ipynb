# 데이터 추출하기

```python
import pandas as pd

df = pd.read_csv("./data/seoul_park03.csv")
df.head()
```

### 조건에 따른 인덱싱 : Boolean indexing

```python
# 어른 입장객 수가 5000보다 큰 날짜들의 데이터 추출
# df[조건]
df[df['어른'] > 5000]
```

```python
# pandas 논리연산자 사용
# 두 개 이상의 조건은 and(&)나 or(|)을 활용하여 같이 사용한다. 이 때 각 조건식은 소괄호()로 묶여있어야 한다.

# 어른 입장객 수가 10000명이 넘는 공휴일의 데이터만을 추출
df[(df['어른']>10000) & (df['공휴일'] == 'O')]
```

```python
# 어른 입장객 수가 10000명이 넘거나 어린이 입장객 수가 2000명이 넘는 날의 데이터 추출
df[(df['어른']>10000) | (df['어린이']>2000)]
```

```python
# 어린이날의 데이터를 추출해 childrensday에 저장
# "월"과 "일" 컬럼을 사용
childrensday = df[(df['월'] == 5) & (df['일'] == 5)]
childrensday
```

### 라벨을 활용한 데이터 추출 : loc

```python
# 2016년 1월 4일의 어른 데이터 추출
# .loc[행의 인덱스, 열의 인덱스]
df.loc[3,"어른"]
```

```python
# 슬라이싱을 활용한 인덱싱
# 유의할점 : loc는 아벨 기반 인덱싱이기 때문에 A:B로 슬라이싱을 하면 A부터 B까지, 즉 B를 포함한 범위를 인덱싱 한다.

df.loc[3:6, "어른":"외국인"]
```

```python
# 어른과 어린이의 입장객 수가 둘 다 1000보다 큰 날짜의 '날짜' 부터 '총계' 행 추출

df.loc[(df['어른'] > 1000) & (df['어린이'] > 1000), '날짜' : '총계']
```

### 순서를 활용한 데이터 추출 : iloc

```python
df.iloc[4,7]
```

```python
# 유의할점 : 위치 기반 인덱싱을 사용하기 때문에 A:B 슬라이싱을 하면 A부터 B-1 까지, 즉 B를 포함하지 않는 범위를 인덱싱한다.
# 3~6번 행, 4~6번 열의 값을 추출

df.iloc[3:7, 4:7]
```

```python
# 특정 위치에 해당하는 값을 추출할 수도 있고, 그 값을 다른 값으로 바꿀수도 있다.
df.loc[4, '청소년'] = 1000
df.head()
```

```python
# 2018년 전체의 날짜, 미세먼지, 초미세먼지 데이터를 변수 dust에 저장
# 2018년 1월 1일의 행 인덱스는 731, 12월 31일의 행 인덱스는 1095 이다.

# 미세먼지 데이터를 불러와 변수 mm에 저장 후 head()를 이용해 데이터 확인
mm = pd.read_csv('./data/misemunji.csv')
mm.head()

# loc 또는 iloc를 활용해서 지정된 데이터를 dust에 저장
dust = mm.loc[731:1095, '날짜':'초미세먼지']    # loc 사용

dust = mm.iloc[731:1096, 0:3]                  # iloc 사용

dust = mm.loc[mm['날짜'].between(20180101,20181231), '날짜':'초미세먼지']   # 날짜 인덱스 값을 모른다는 가정
dust
```

```python
mm['날짜'] = pd.to_datetime(mm['날짜'], format="%Y%m%d")    # datetime으로 날짜 컬럼을 포맷
mm.head()

# dust = mm.loc[mm['날짜'].dt.year == 2018, '날짜':'초미세먼지']

dust = mm.loc[(mm['날짜']>= "2018-01-01") & (mm['날짜'] <= "2018-12-31"), '날짜':'초미세먼지']  # 대소비교

dust
```
