# Matplotlib 기본 문법

```python
import matplotlib.pyplot as plt

import koreanize_matplotlib
```

```python
# matplotlib 데이터
year = [2018, 2019, 2020, 2021, 2022, 2023]
python = [5.8, 8.17, 9.31, 11.87, 15.63, 14.51]
C = [13.59, 14.08, 16.72, 14.32, 12.71, 14.73]
java = [15.78, 15.04, 16.73, 11.23, 10.82, 13.23]
JS = [3.49, 2.51, 2.38, 2.44, 2.12, 2.1]
```

```python
plt.title("프로그래밍 언어 점유율", fontsize=25)   # 그래프 제목 설정
plt.xlabel("연도", fontsize=15)      # x축 레이블 설정
plt.ylabel("점유율", fontsize=15)    # y축 레이블 설정

# 선 그래프 : plot(x축, y축)
plt.plot(year,python, label='python')

# 다른 그래프를 추가할 때 여러번 plot()호출
plt.plot(year, C, label='C')
plt.plot(year, java, label='Java')
plt.plot(year, JS, label='JS')
plt.legend()    # plot안에 label로 이름을 지정해준 것을 그대로 가져옴

# plt.legend(['python', 'C', 'java', 'JS'])   # 그래프를 그린 순서대로 정의 추천하는 방법은 아님
```

## figure

```python
fig1 = plt.figure(1,figsize=(4,3))      # id가 1인 새로운 figure 객체 생성
plt.plot(year, python, label='Python')  # 현재 선택된 figure에 선 그래프를 그린다.
plt.title('fig1')

fig2 = plt.figure(2,figsize=(4,3))
plt.plot(year,java, label='Java')
plt.title('fig2')

plt.figure(1)   # id가 1인 figure 선택
plt.plot(year,C, label='C')     # 현재 선택된 figure에 선 그래프를 그린다.
```

## subplot
하나의 figure에 여러 개의 그래프를 그리기 위해 사용

```python
fig, (ax1, ax2) = plt.subplots(1,2) # subplot(행,열) 세로 1칸, 가로 2칸인 총 2개의 subplot 생성
# 생성된 서브플롯은 각각 ax1, ax2 에 저장된다.

ax1.plot(year, python)
ax2.plot(year, java)

plt.show()  # 그래프를 표시(주피터 환경에서는 생략 가능)
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, constrained_layout=True)    # (행,열), constrained_layout=True : 서브플롯 그래프간의 간격조절

ax1.plot(year, python) 
ax1.set_title('ax1의 제목')     # subplot의 제목을 정할 때 사용
ax1.set_xlabel('x라벨')

ax2.plot(year, java)
ax2.set_title('ax2의 제목')

plt.show()
```

아래의 코드를 완성하여 다음을 만족하는 그래프를 그리세요.

- `plt.subplots()` 함수를 이용해 1행 2열로 총 두개의 서브플롯을 생성합니다.
- 생성된 서브플롯은 변수 `ax1`, `ax2`에 저장합니다.
- 서브플롯 ax1
    - 파이썬의 연도별 점유율을 선 그래프로 그립니다.
    - 제목: "파이썬의 점유율"
    - x축 레이블: "연도"
    - y축 레이블: "점유율"
- 서브플롯 ax2
    - 자바의 연도별 점유율을 선 그래프로 그립니다.
    - 제목: "자바의 점유율"
    - x축 레이블: "연도"

```python
fig3, (ax1,ax2) = plt.subplots(1,2, constrained_layout=True)

ax1.plot(year, python)
ax1.set_title('파이썬의 점유율')
ax1.set_xlabel('연도')
ax1.set_ylabel('점유율')

ax2.plot(year, java)
ax2.set_title('자바의 점유율')
ax2.set_xlabel('연도')

plt.show()
```
