# 산점도와 히트맵
산점도와 히트맵은 데이터의 상관관계를 볼 때 유용한 그래프이다.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import koreanize_matplotlib
```

```python
df = pd.read_csv('seoul_park.csv')
df.head()
```

## 산점도
plt.scatter()

```python
# 어른 입장객 수와 청소년 입장객 수의 산점도 그래프
plt.scatter(df['어른'],df['청소년'],)
plt.show()
```

```python
# 어른과 어린이 입장객 수의 산점도 그래프
plt.scatter(df['어른'],df['어린이'])
plt.show()
```

```python
# seaborn을 이용한 산점도 그래프
sns.scatterplot(data=df, x='어른', y='어린이')

plt.ylim(0,6000)    # y축 범위를 0에서 6000으로 제한
plt.xlim(0,30000)   # x축 범위를 0에서 30000으로 제한
plt.show()
```

## 히트맵
##### 데이터 값의 상대적인 크기를 색상으로 표시하여 시각화 하는 그래프이다.
##### 데이터의 밀도, 페턴, 상관관계 등을 시각적으로 파악할 수 있다.

```python
# 피어슨 상관계수   //  -1 ~ 1 의 범위를 갖는 수치 // -1에 가까워질수록 음의 상관관계 , 1에 가까워질수록 양의 상관관계 , 0이면 아무련 관련이 없는 관계
df_corr = df.corr(numeric_only=True)
df_corr
```

```python
# 히트맵
# 위의 상관계수 행렬을 이용하여 히트맵을 그린다.
sns.heatmap(df_corr, cmap='coolwarm')
```
