# Seaborn 기본 문법

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("dark")
import koreanize_matplotlib
```

```python
# matplotlib 데이터
year = [2018, 2019, 2020, 2021, 2022, 2023]
python = [5.8, 8.17, 9.31, 11.87, 15.63, 14.51]
C = [13.59, 14.08, 16.72, 14.32, 12.71, 14.73]
java = [15.78, 15.04, 16.73, 11.23, 10.82, 13.23]
JS = [3.49, 2.51, 2.38, 2.44, 2.12, 2.1]
```

## 그래프 그리기

```python
sns.lineplot(x=year, y=python)
```

```python
# matplotlib 기반으로 개발된 라이브러리이기 때문에 matplotlib 매개변수 대부분을 사용 가능
sns.lineplot(x=year, y=python, label='python')
sns.lineplot(x=year, y=java, label='Java')
```

```python
# sns.lineplot()은 matplotlib.axes를 반환하는 함수이다.
# matplotlib.axes 메소드를 이용해 그래프의 모양을 조절할 수 있다.
lp = sns.lineplot(x=year, y=python, label='python')

lp.set_xlabel('년도')
lp.set_ylabel('점유율')
```

## 테마 사용하기

```python
# sns.set_style("whitegrid")
plt.title('김성호')
lp = sns.lineplot(x=year, y=python, label='Python')
```

## subplot()

```python
fig, (ax1,ax2) = plt.subplots(1,2, constrained_layout=True)

fig = sns.lineplot(x=year, y=python, label='Python', ax=ax1)    # matplotlib와는 다르게 ax=으로 subplot을 설정해 줄 수 있다.
sns.lineplot(x=year, y=java, label='Java', ax=ax2)
```

- `plt.subplots()` 함수를 이용해 1행 2열로 총 두개의 서브플롯을 생성합니다.
- 생성된 서브플롯은 변수 `ax1`, `ax2`에 저장합니다.
- 서브플롯 ax1
    - `sns.lineplot()`을 이용하여 `C`언어의 연도별 점유율을 선 그래프로 그립니다.
    - 처음 정의한 데이터에서 `C`를 사용합니다.
- 서브플롯 ax2
    - `sns.lineplot()`을 이용하여 `JS`의 연도별 점유율을 선 그래프로 그립니다.
    - 처음 정의한 데이터에서 `JS`를 사용합니다.

```python
fig, (ax1, ax2) = plt.subplots(1,2, constrained_layout=True)

sns.lineplot(x=year, y=C, ax=ax1)
sns.lineplot(x=year, y=JS, ax=ax2)
```
