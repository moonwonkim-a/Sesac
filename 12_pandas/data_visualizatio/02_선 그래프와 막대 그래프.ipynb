# 선 그래프와 막대 그래프

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams["axes.unicode_minus"] = False
plt.rc("font", family="Malgun Gothic")
```

```python
import pandas as pd

df = pd.read_csv("seoul_park.csv")
df.head()
```

```python
df.info()
```

### 1. 선 그래프

#### 1-1 Matplotlib

월별 데이터 묶기

```python
gb_mon = df.groupby('월')
```

```python
# 월별 평균 데이터

df_mon = gb_mon.mean(numeric_only=True)     # numeric_only=True : 숫자형 열만 평균 계산
df_mon
```

```python
# 어른, 어린이, 청소년별 월별 입장객수 선그래프로 보기
plt.title('서울대공원 입장객 분석', fontsize=20)    # title('그래프 제목')'
plt.plot(df_mon['어른'],label="어른")
plt.plot(df_mon['어린이'],label="어린이")
plt.plot(df_mon['청소년'],label="청소년")
plt.xlabel('월', fontsize=15)
plt.ylabel('입장객 수',fontsize=15)
plt.legend()
plt.show()
```

```python
# 어른, 어린이, 청소년별 월별 입장객수 선그래프로 보기
plt.title('서울대공원 입장객 분석', fontsize=20)    # title('그래프 제목')
plt.plot(df_mon['어른'],label="어른")
plt.plot(df_mon['어린이'],label="어린이")
plt.plot(df_mon['청소년'],label="청소년")
plt.xlabel('월', fontsize=15)
plt.ylabel('입장객 수',fontsize=15)
plt.legend()
plt.xticks([mon for mon in range(1, 13)])   # x축 눈금 조절 : 1부터 12가 들어있는 리스트를 xticks에 전달
plt.show()
```
