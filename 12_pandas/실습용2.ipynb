```python
import pandas as pd
import numpy as np
```

```python
employee_data = {
    'EmployeeID': [101, 102, 103, 104, 105],
    'Name': ['John', 'Emily', 'Sara', 'Tom', 'Anna'],
    'Department': ['HR', 'IT', 'Marketing', None, 'HR'],
    'Salary': [50000, 70000, None, 45000, 52000],
    'YearsAtCompany': [5, None, 8, 3, 4]
}

em = pd.DataFrame(employee_data)
em.head()
```

```python
em_fillna = ['Salary', 'YearsAtCompany']

for emf in em_fillna:
    em[emf] = em[emf].fillna(em[emf].mean())
em
```

```python
product_data = {
    'ProductID': [1, 2, 3, 1, 4, 2],
    'ProductName': ['Laptop', 'Mouse', 'Monitor', 'Laptop', 'Keyboard', 'Mouse'],
    'Price': [1200, 25, 300, 1150, 45, 22],
    'QuantitySold': [10, 100, 50, 8, 70, 120]
}
pdd = pd.DataFrame(product_data)
pdd.head()
```

```python
pdd = pdd.sort_values('Price').drop_duplicates('ProductID', keep='first')
pdd
```

```python
car_data = {
    'Model': ['Sedan', 'SUV', 'Truck', 'Coupe', 'Convertible'],
    'Horsepower': [180, 250, 200, 170, 150],
    'MPG': [30, 20, 15, 35, 28],
    'Price': [30000, 45000, 40000, 28000, 32000]
}
cd = pd.DataFrame(car_data)
cd
```

```python
cd = cd.sort_values('Horsepower').head(3).sort_values('MPG',ascending=False)
cd
```
